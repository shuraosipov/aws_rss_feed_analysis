```python
import pandas as pd
```

```python
df = pd.read_csv('aws_release_notes.csv')
```

```python
df
```

```python
# drop rows with NaN values
df = df.dropna()
```

```python
df
```

```python
df['services']
```

```python
# remove spaces from the beginning and end of the string in the 'services' column
df['services'] = df['services'].str.strip()
```

```python
# split cell into multiple rows based on the space separator
df3 = df.assign(services=df.services.str.split(' ')).explode('services')
```

```python
df3
```

```python
# save dateframe to csv
df3.to_csv('aws_release_notes_cleaned.csv', index=False, sep=';')
```
